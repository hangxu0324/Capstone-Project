launch_date: Date when the brand was launched in the market. Approval timelines are different in each region.

```python
launch_df_anon = pd.read_csv('../data/launch_df_anon.csv').drop(['Unnamed: 0', 'Unnamed: 0.1'], axis = 1)
launch_df_anon
```

```python
df = board_anon.merge(cx_df_anon, on = ['country', 'cluster_id'], how = 'outer') \
.merge(gx_df_anon, on = ['country', 'cluster_id'], how = 'inner') \
.merge(launch_df_anon, on =  'cluster_id', how = 'inner')
df.head()
```

```python
df['stage_lag_combined'] = df.apply(lambda x: str(x.stage_name) + '_' + str(x.stage_name_lag_1), axis = 1)
df.stage_lag_combined.value_counts()
```
